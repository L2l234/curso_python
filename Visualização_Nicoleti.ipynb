```python

#importação do arquivo
from google.colab import files
import pandas as pd

uploaded = files.upload()

nome_do_arquivo = list(uploaded.keys())[0]

dados=pd.read_csv(nome_do_arquivo)

print(dados.head())
```

```python
#estatísticas descritivas resumidas
estatisticas_descritivas = dados.describe()
print(estatisticas_descritivas)
```

```python
#estatísticas descritivas de todas as colunas
estatisticas_descritivas = dados.describe(include='all')
print(estatisticas_descritivas)
```

```python
#gráfico violino usando a Coluna Sex e a coluna Salary(M-Homem, F-mulher)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.violinplot(x='Sex', y='Salary', data=dados)
plt.title('Distribuição de Salários por Gênero')
plt.show()
```

```python
#testando a coluna GenderID no lugar de Sex,o resultado foi igua(1-homem, 0-mulher)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 16))
sns.violinplot(x='GenderID', y='Salary', data=dados)
plt.title('Distribuição de Salários por Gênero')
plt.show()
```

```python
dados = pd.read_csv(nome_do_arquivo, parse_dates=['DateofHire'], dayfirst=True)
print(dados.head)
```

```python
#Série temporal(fiquei com um pouco de dificuldade em definir os tamanhos dos gráficos)
plt.figure(figsize=(10, 6))
dados['Ano_Mes'] = dados['DateofHire'].dt.to_period('M')
dados.groupby('Ano_Mes').size().plot(marker='o')
plt.title('Número de Contratações ao Longo do Tempo')
plt.xlabel('Ano e Mês')
plt.ylabel('Número de Contratações')
plt.show()
```

```python
#histograma de salários
plt.figure(figsize=(10, 6))
sns.histplot(dados['Salary'], bins=100, kde=True, color='skyblue')
plt.title('Histograma de Salários')
plt.xlabel('Salário')
plt.ylabel('Frequência')
plt.show()
```
